```python
import numpy as np
arr1 = np.array([1, 2, 3]) # создание numpy массива из обычного
zeros = np.zeros((2, 3)) # Матрица 2 на 3, заполненная нулями
ones = np.ones((2, 3)) # Матрица 2 на 3, заполненная единицими
arr2 = np.arange(10, 51, 10) # Создает последовательность [10; 51) с шагом 10
```

```python
print(zeros.shape) # Вывод формы матрицы zeros
print(zeros.dtype) # Вывод типа данных, хранимых в zeros
```

```python
arr1 = [10, 20, 30, 40]

# Индексирование — получение одного элемента
print(arr1[0])  # вывод: 10

# Срез — получение подмассива
print(arr1[0:2])  # вывод: [10, 20]
```

```python
print(arr1 + ones) # Поэлементное сложение
print(arr1.mean()) # Средняя величина
```
